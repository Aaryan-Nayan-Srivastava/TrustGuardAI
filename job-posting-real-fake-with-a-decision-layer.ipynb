```python
drop_col = [
    "job_id",
    "location",
    "title",
    "company_profile",
    "description",
    "requirements",
    "benefits",
    "telecommuting",
    "function",
    "industry"
]

df = data.drop(columns=drop_col)
```

```python
df.head()
```

```python
allowed_departments = {
    "sales",
    "engineering",
    "marketing",
    "operations",
    "development",
    "it",
    "product"
}


df["department"] = (
    df["department"]
    .fillna("unknown")
    .astype(str)
    .str.lower()
    .str.strip()
)

df["department"] = df["department"].apply(
    lambda x: x if x in allowed_departments else "other"
)

print(df["department"].value_counts(normalize=True))
```

```python
df.head()
```

```python
df["salary_range"] = (
    df["salary_range"]
    .fillna("unknown")
    .astype(str)
    .str.strip()
    .str.lower()
)

invalid_salary_values = {"unknown", "0-0", "0", ""}

df["has_salary_range"] = (
    ~df["salary_range"].isin(invalid_salary_values)
).astype(int)
df = df.drop(columns=["salary_range"])
```

```python
df["employment_type"] = (
    df["employment_type"]
    .fillna("unknown")
    .astype(str)
)
```

```python
df.sample(5)
```

```python
df["required_experience"] = (
    df["required_experience"]
    .fillna("missing")
    .astype(str)
    .str.lower()
    .str.strip()
)
```

```python
df["experience_missing"] = (
    df["required_experience"] == "missing"
).astype(int)

df["experience_not_applicable"] = (
    df["required_experience"] == "not applicable"
).astype(int)
```

```python
def map_experience(x):
    if x in ['internship']:
        return 'internship'
    if x in ["entry level", "associate"]:
        return "junior"
    if x in ["mid-senior level", "director", "executive"]:
        return "senior"
    return "unspecified"

df["required_experience"] = df["required_experience"].apply(map_experience)
```

```python
df["required_education"] = (
    df["required_education"]
    .fillna("missing")
    .astype(str)
    .str.lower()
    .str.strip()
)


df["education_missing"] = (
    df["required_education"] == "missing"
).astype(int)

df["education_unspecified"] = (
    df["required_education"] == "unspecified"
).astype(int)

def map_education(value):
    if "bachelor" in value:
        return "bachelors"
    if "master" in value:
        return "masters"
    if "high school" in value:
        return "high School or equivalent"
    if value in ["missing", "unspecified"]:
        return "unspecified"
    return "other"

df["required_education"] = df["required_education"].apply(map_education)
```

```python
x = df.drop(columns=["fraudulent"])
y = df["fraudulent"]

x = pd.get_dummies(x, drop_first=True)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.3,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000,
    class_weight="balanced"
)

model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=(y_train == 0).sum() / (y_train == 1).sum(),
    eval_metric="logloss",
    random_state=42
)

xgb.fit(x_train, y_train)
```

```python
y_xgb=xgb.predict(x_test)
```

```python
print(accuracy_score(y_test, y_xgb))
print(confusion_matrix(y_test, y_xgb))
```

```python
import numpy as np

p = xgb.predict_proba(x_test)[:, 1]

df_meta = pd.DataFrame({
    "predicted_label": (p >= 0.5).astype(int),
    "predicted_probability": p,
    "confidence_margin": np.abs(p - 0.5),
    "entropy": -(p * np.log(p + 1e-8) + (1 - p) * np.log(1 - p + 1e-8)),
    "true_label": y_test.values
})
```

```python
df_meta.head()
```

```python
df_meta["is_correct"] = (
    df_meta["predicted_label"] == df_meta["true_label"]
).astype(int)

entropy_threshold = df_meta["entropy"].quantile(0.75)

df_meta["low_confidence"] = (
    df_meta["entropy"] >= entropy_threshold
).astype(int)


df_meta["needs_human_review"] = (
    (df_meta["is_correct"] == 0) |
    (df_meta["low_confidence"] == 1)
).astype(int)
```

```python
df_meta.head()
```

```python
df_meta_B = df_meta.drop(columns=["true_label", "is_correct", "low_confidence"])
```

```python
df_meta_B.sample(5)
```

```python
df_meta_B.shape
```

```python
df_meta_B["needs_human_review"].value_counts(normalize=True)
```

```python

X_B = df_meta_B.drop(columns=["needs_human_review"])
y_B = df_meta_B["needs_human_review"]

X_B_train, X_B_val, y_B_train, y_B_val = train_test_split(
    X_B,
    y_B,
    test_size=0.3,
    stratify=y_B,
    random_state=42
)
```

```python
from xgboost import XGBClassifier

model_B = XGBClassifier(
    n_estimators=120,
    max_depth=3,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    random_state=42
)

model_B.fit(X_B_train, y_B_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_B_prob = model_B.predict_proba(X_B_val)[:, 1]
y_B_pred = (y_B_prob >= 0.5).astype(int)

print(confusion_matrix(y_B_val, y_B_pred))
print(classification_report(y_B_val, y_B_pred))
```

```python
accuracy_score(y_B_val,y_B_pred)
```

```python
import numpy as np

for t in [0.3, 0.4, 0.5, 0.6]:
    preds = (y_B_prob >= t).astype(int)
    fn = ((preds == 0) & (y_B_val == 1)).sum()
    print(f"Threshold {t}: False negatives = {fn}")
```
